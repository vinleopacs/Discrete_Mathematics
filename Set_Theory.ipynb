   - This identity shows that the set difference A\B is equal to the intersection of A and the complement of B. The proof establishes that both sides are subsets of each other.

3. **B\A = B $\cap$ $A^{c}$**
   - This identity indicates that the set difference B\A is equal to the intersection of B and the complement of A. The proof establishes the equality of both sides.

4. **A $\cup$ B = A $\cup$ (B $\cap$ $A^{c}$)**
   - This identity shows that the union of sets A and B is equal to the union of A and the intersection of B and the complement of A. The proof confirms their equality.

5. **$(\bigcup_{i=1}^{\infty} A_{i}^{c})^{c}$ = $\bigcap_{i=1}^{\infty}A_{i}$**
   - This identity leverages De Morgan's law to establish that the complement of the union of the complements of a collection of sets is equivalent to the intersection of the original sets.

6. **|A $\cup$ B| = |A| + |B| - |A $\cap$ B|**
   - This identity is known as the inclusion-exclusion principle. It provides a formula for calculating the size of the union of two sets by considering their individual sizes and the size of their intersection.

These proofs demonstrate the equivalence of the given set identities, providing a deeper understanding of set theory and relationships between sets.

```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn3

# Define the sizes of each set
set_sizes = (65, 45, 42, 20, 25, 15, 8)

# Create a Venn diagram
venn = venn3(subsets=set_sizes, set_labels=('Newsweek', 'Time', 'Fortune'))

# Calculate the number of people who read at least one magazine
at_least_one = sum(set_sizes) - int(venn.get_label_by_id('100').get_text().replace('\n', ''))
print("Number of people who read at least one magazine:", at_least_one)

# Find the number of people who read exactly one magazine
exactly_one = (
    int(venn.get_label_by_id('100').get_text()) +
    int(venn.get_label_by_id('010').get_text()) +
    int(venn.get_label_by_id('001').get_text())
)
print("Number of people who read exactly one magazine:", exactly_one)

plt.title("Venn Diagram of Magazine Readers")
plt.show()
```
